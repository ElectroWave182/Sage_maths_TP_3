### Exercice n° 2 : Faire son histogramme



On donne une liste de notes, par exemple :

    notes=[4, 15, 15, 17, 20, 20, 8, 3, 8, 10, 11, 14, 14, 12, 3]
et on demande de tracer l'histogramme de ces notes, comme ci-dessous :

![ ](histo.png)

On utilisera la fonction `line2d`.

On pourra utiliser la méthode `count` d'une liste qui renvoie le nombre d'occurrence d'un élément dans une liste. Par exemple, `L.count(42)` renvoie le nombre d'occurrences de 42 dans la liste `L`.

Pour extraire les éléments distincts d'une liste, on utilisera le constructeur `set` : si `L` est une liste alors `list(set(L))` est formé de la liste des éléments distincts de `L`.



```python
notes = [4, 15, 15, 17, 20, 20, 8, 3, 8, 10, 11, 14, 14, 12, 3]
l, sortie = list(set(notes)), []
for ele in l:
    sortie.append(line2d([(ele, 0), (ele, notes.count(ele))], thickness = 4))
show(sum(sortie))
```

### Exercice n° 3 : Le graphe de Petersen



Le graphe ci-dessous est appelé le graphe de Petersen.

![ ](petersen.png)

Le reproduire.



```python
arg = pi / 10
points = []
sortie = []
for i in range(1, 11):
    mult = (i - 1) // 5 + 1
    coord = (cos(arg) * mult, sin(arg) * mult)
    points.append(coord)
    sortie.append(point(coord, color = "red", size = 300, zorder = 3))
    sortie.append(text(i, coord, color = "black", fontsize = 14, zorder = 3))
    arg += 2 * pi / 5
for i in range(10):
    for j in range(10):
        if i % 5 == j % 5:
            sortie.append(line2d([points[i], points[j]]))
        if i < 5 and j < 5:
            sortie.append(line2d([points[i], points[j]]))
        if j - 1 == i and i > 4:
            sortie.append(line2d([points[i], points[j]]))
sortie.append(line2d([points[5], points[9]]))
show(sum(sortie), xmin = -5, xmax = 5, ymin = -3, ymax = 3, axes = False)
```
